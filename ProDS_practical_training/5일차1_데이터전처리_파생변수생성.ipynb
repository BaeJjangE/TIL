### 파생변수 : 기존 변수를 조합하여 새로운 변수를 만들어냄

### 파생변수의 예시
- 기온, 습도, 풍속을 조합하여 만든 체감온도 변수
- 물건 주문 건수와 환불 건수를 조합하여 만든 환불 비율 변수
- 기존 방문 매장 정보를 활용한 주 방문 매장 변수

### numpy - where(): if랑 비슷
- 조건에 따라 두 개의 출력을 내는 함수
- if() 함수를 대체할 수 있으며 조건, True일 때 반환값, False일 때 반환값을 차례대로 기입

### pandas - rename(): 데이터프레임의 변수명을 변경할 때 사용하는 메서드
- columns 인자에 기존 변수명과 신규 변수명의 쌍을 딕셔너리로 구성하여 입력

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df["is_setosa"] = np.where(df["Species"] == "setosa", 1, 0)
df.head()
```

```python
df["is_setosa2"] = (df["Species"] == "setosa") + 0
df["is_setosa3"] = (df["Species"] == "setosa") * 1
df.head()
```

```python
pd.crosstab(df["Species"], df["is_setosa2"]) # 잘 들어갔는지 교차확인
```

```python
df = df.rename(columns = {"Sepal.Length": "S.L", "Sepal.Width": "S.W"})
df
```

### pandas - get_dummies()
 - 편리한 가변수 생성(One Hot Encoding)을 도와주는 함수
 - columns 인자에 명목형 변수 지정 및 처리 가능
 - drop_first 인자에 True를 입력하면 마지막 가변수 제외 후 생성

```python
bike.head(2)
```

```python
bike_dum = pd.get_dummies(data = bike, columns = ["season"]) # 하나인데도 대괄호로 묶어줘야하는 함수들이 있음
bike_dum.head(2)
```

```python
bike_dum = pd.get_dummies(data = bike, columns = ["season"], drop_first = True) # drop_first = True 는 첫번째 더미를 삭제, 
# 그 이유는 완전공선성 문제 발생을 제지하기위함
bike_dum.head(2)
```

## 문제 2: casual 값의 최대값이 25가 넘은 날은 총 며칠인가?
- bike.csv 파일사용
- date 어트리뷰트 활용


```python
bike = pd.read_csv("bike.csv")
bike.head(20)
```

```python
bike["datetime"] = pd.to_datetime(bike["datetime"])
bike.head(2)
```

```python
bike["date"] = bike["datetime"].dt.date
bike.head()
```

```python
bike_agg = bike.groupby("date")["casual"].max() # "date" 컬럼인덱스 겹치는 값 중 "casual" 값이 max인 값
bike_agg
```

```python
a = bike_agg > 25
bike_agg = bike_agg[a]
```

```python
bike_agg
```

```python
type(bike_agg) # 시리즈이므로 iloc, loc 안써짐
```

```python
bike_agg_up25 = bike_agg[bike_agg > 25] # 바로 아래거 참고
bike_agg_up25.head()
```

```python
bike_agg > 25 # 참고자료
```

```python
len(bike_agg_up25)
```

데이터프레임으로 핸들링하는 방법

```python
bike_agg = bike.groupby("date")["casual"].max().reset_index() # reset_index(): 시리즈를 자동으로 데이터프레임으로 변경
bike_agg.head()
```

```python
len(bike_agg.loc[bike_agg["casual"] > 25, ])
```

```python
bike_agg2 = bike.groupby(["date", "temp"])["casual", "humidity"].max()
print(bike_agg2)
type(bike_agg2)
```

#### 객체.groupby() : 괄호안에 들어온 컬럼인덱스를 기준으로 데이터를 변환해줌

## 문제3: 시간대별 registered 평균을 산출했을 때 값이 가장 큰 시간은?
- bike.csv 파일 사용
- hour 어트리뷰트 활용

```python
bike = pd.read_csv("bike.csv")
bike.head()
```

```python
bike_time = pd.to_datetime(bike["datetime"])
bike["hour"] = bike_time.dt.hour
bike
```

```python
a = bike.groupby("hour")["registered"].mean().reset_index()
a
```

```python
a.loc[a["registered"] == a["registered"].max(), ]
```

```python
# 다른방법
a.loc[a["registered"].idxmax(), ] # idxmax : 최대값 위치찾기
```
