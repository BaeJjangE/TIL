### 이상치의 처리 방법
- 이상치 처리에 절대적인 기준은 없음
- 대표적으로 Carling, Tukey 방법이 있음
- 분포 기반으로 처리도 가능

### 결측치: 값이 기록되지않고 비어있음

### 결측치의 처리 방법
- 결측치 처리에 절대적인 기준은 없음
- 단순 제거와 특정 값으로 대체하는 방식이 있음
 - 대체는 평균값이나 중앙값으로 대체
- 분석 데이터에서 결측치가 차지하는 비중이 낮은 경우 단순 제거하는 경우가 많음
 - 1%나 0.1% 정도 일 경우 사용 가능

### pandas - dropna()
- 결측치가 있는 row또는 column을 제거하기 위함
- how인자에 'any'는 결측치가 하나라도 있는 경우, 'all'은 전체가 결측인 경우 제거

### pandas - quantile()
- 분위수를 연산하는 메서드로 이상치 필터링에 활용
- q 인자에 1개 이상의 확률값(0~1)을 입력하여 그에 맞는 분위수 산출 가능

```python
df_na2 = df.head(7)
df_na2
```

```python
df_na2.dropna() # 결측치가 하나라도 있으면 그 row자체를 없애버림, 디폴트값: how = 'any'
```

```python
df_na2.dropna(how = "all") # 모두 결측치여야 삭제한다는 뜻
```

```python
df_na2.iloc[:, :-1].dropna(how = "all")
```

### 평균을 기준으로 1.5 표준편차를 넘어서는 값을 이상치라고 간주 할 때 Sepal.Length 변수를 기준으로 이상치인 row 개수는 몇개인가?
- iris.csv

```python
df = pd.read_csv("iris.csv")
df
```

```python
sl_mean = df["Sepal.Length"].mean()
sl_std = df["Sepal.Length"].std()
print(sl_mean)
print(sl_std)
```

```python
cond_1 = df["Sepal.Length"] < (sl_mean - 1.5 * sl_std)
print(cond_1)
cond_2 = df["Sepal.Length"] > (sl_mean + 1.5 * sl_std)
print(cond_2)
df_out = df.loc[(cond_1) | (cond_2), ] # '|'  'or'이라는 뜻
print(df.loc[(cond_1) | (cond_2), ])
len(df_out)
```
