## 병합의 종류
- 바인딩(binding)[concat()]은 행 또는 열 방향으로 단순 붙이기
- 조인(join)[merge()]은 특정변수를 기준으로 엮기
 - 조인 연산 중 Inner Join과 Left Join을 많이 사용


### pandas - reset_index()
 - 기존의 인덱스를 초기화하는 메서드
 - 데이터프레임의 인덱스를 초기화 할 경우 0부터 row개수 만큼의 숫자가 재할당
 - groupby() 등 어떠한 연산 이후에 다시 리셋시킬때 사용
### pandas - set_index()
 - 특정 변수를 인덱스로 지정할 경우 사용하는 메서드
 - 데이터병합 또는 시계열 분해에서 연산을 위해 사용

### pandas - concat()
 - 데이터를 row 방향 또는 column 방향으로 이어붙이는 함수
 - axis인자에 1을 입력하는 경우 column 방향으로 병합
 - 데이터 병합 시 서로 다른 변수 속성 또는 인덱스에 주의

```python
bike_1 = bike.iloc[:3, :4]
bike_2 = bike.iloc[5:8, :4]
print(len(bike_1))
print(len(bike_2))
```

```python
bike_1
```

```python
bike_2
```

```python
pd.concat([bike_1, bike_2]) # []로 묶어 줘야함, row 방향
```

```python
pd.concat([bike_1, bike_2], axis = 1) # 옆으로 붙이기
```

```python
pd.concat([bike_1, bike_2.reset_index(drop = True)], axis = 1)
```

```python
pd.concat([bike_1.reset_index(drop = True), bike_2.reset_index(drop = True)], axis = 1) # 둘 다 번호 초기화시켜서 합침
```

혼자 실습해보기 - 행 인덱스 수가 다를경우

```python
bike_1 = bike.iloc[:3, :4]
bike_2 = bike.iloc[5:9, :4]
print(len(bike_1))
print(len(bike_2))
```

```python
pd.concat([bike_2.reset_index(drop=True), bike_1], axis = 1)
```

### pandas - merge() 
 - 데이터프레임을 변수 기준으로 병합할 때 사용하는 함수
 - inner join, left join 등 다양한 join 연산이 가능하며 비슷한 함수로는 join이 있음
 - left_on과 right_on 인자에 각 데이터의 key 변수를 지정

```python
df_A = pd.read_csv("join_data_group_members.csv")
df_B = pd.read_csv("join_data_member_room.csv")
```

```python
df_A.head(2)
```

```python
df_B.head(2)
```

```python
pd.merge(left = df_A, right = df_B,
        left_on = "member", right_on = "name",
        how = "inner")
```

```python
pd.merge(left = df_A, right = df_B,
        left_on = "member", right_on = "name",
        how = "left")
```

## 2. 여름과 겨울의 시간대별 registered 평균을 비교할 때 가장 차이가 많이나는 시각은?
- bike.csv 파일 사용
- hour 어튜리뷰트 활용

```python
bike = pd.read_csv("bike.csv")
bike.head()
```

```python
bike_2 = (bike["season"] == 2)
bike_4 = (bike["season"] == 4)
```

```python
bike_2 = bike[bike_2]
bike_4 = bike[bike_4]
```

```python
bike_2.head()
```

```python
bike_dt2 = pd.to_datetime(bike_2["datetime"])
bike_dt4 = pd.to_datetime(bike_4["datetime"])
```

```python
bike_2
```

```python
bike_2["date"] = bike_dt2.dt.time
bike_4["date"] = bike_dt4.dt.time
```

```python
bike_2.head()
```

```python
mg_mean2 = bike_2.groupby("date")["registered"].mean()
mg_mean4 = bike_4.groupby("date")["registered"].mean()
```

```python
sub = abs(mg_mean2 - mg_mean4)
```

```python
sub.idxmax()
```

정답

```python
bike = pd.read_csv("bike.csv")
bike.head()
```

```python
bike['datetime'] = pd.to_datetime(bike["datetime"])
bike["hour"] = bike["datetime"].dt.hour
```

```python
bike_s2 = bike.loc[bike["season"] == 2, ]
bike_s4 = bike.loc[bike["season"] == 4, ]
print(bike_s2["season"].unique()) # 머가 들어있는지 확인
print(bike_s4["season"].unique()) # 머가 들어있는지 확인
```

```python
bike_s2_agg = bike_s2.groupby("hour")["registered"].mean().reset_index()
bike_s4_agg = bike_s4.groupby("hour")["registered"].mean().reset_index()
bike_s2_agg.head()
```

```python
bike_agg_bind = pd.concat([bike_s2_agg, bike_s4_agg], axis = 1)
bike_agg_bind.head(2)
```

```python
bike_agg_bind = bike_agg_bind.iloc[:, [0, 1, 3]]
bike_agg_bind.head(2)
```

```python
bike_agg_bind.columns = ["hour", "reg_s2", "reg_s4"]
bike_agg_bind.head(2)
```

```python
bike_agg_bind["diff"] = bike_agg_bind["reg_s2"] - bike_agg_bind["reg_s4"]
bike_agg_bind.head(2)
```

```python
bike_agg_bind.iloc[[bike_agg_bind["diff"].idxmax()], ]
```

```python
bike_agg_bind["diff"].idxmax()
```

```python
bike = pd.read_csv("bike.csv")
bike.head(2)
```

```python
bike["datetime"] = pd.to_datetime(bike["datetime"])
bike["date"] = bike["datetime"].dt.date
bike.head(2)
```

```python
bike_h100 = bike.groupby(["date"])["humidity"].max().reset_index()
bike_h100
```

```python
bike_h100 = bike_h100.loc[bike_h100["humidity"] == 100, ]
bike_h100
```

```python
len(bike_h100)
```

```python
bike
```

```python
bike_join = pd.merge(left = bike, right = bike_h100, left_on ="date", right_on = "date", how = "inner")
bike_join
```

```python
bike_join_temp_up30 = bike_join[bike_join["temp"] > 30]
bike_join_temp_up30
```

```python
bike_join_temp_up30["count"].mean()
```
