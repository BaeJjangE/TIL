## Key words
### 평균비교, ANOVA, 이원분산분석, ols, statsmodels, anova_lm

## 이원 분산 분석의 특징
- 수치형 종속변수 1개와 명목형 독립변수가 2개일 때 실시하는 분석
 - ols모델에 명목형 변수이면서 수치형일 경우는 C() 넣어주는거 잊지말기
- 두 요인이 서로 상호간에 영향을 주고 받으면서 나타나는 교호작용 효과(interaction effect) 확인가능

### 가설 - 주요 효과(main effect)
- 귀무가설 : 집단간 평균이 같음
- 대립가설 : 평균이 같지 않은 집단이 한쌍 이상 존재

### 가설 - 교호작용 효과(interaction effect)
- 귀무가설 : 요인 간 교호작용이 없음
- 대립가설 : 요인 간 교호작용이 있음

### 이원 분산 분석 - ols()
 - statsmodels의 분산 분석을 수행하는 함수
 - 이원 분산 분석에서는 각 요인의 교호작용을 확인하기 위해서 ':'를 사용

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_csv("diamonds.csv")
df.head(2)
```

cut, color, price로 해봄

```python
formula = "price ~ cut + color + cut:color"
model = ols(formula, data = df).fit()
anova_lm(model)
```

- 주요효과(cut+color), 교호작용효과(cut:color)가 나옴
- 교호작용효과: PR을 보면 귀무가설 기각하고 대립가설을 채택

### 1. 다음 중 종속변수를 registered, 독립변수를 season, holiday로 지정했을 때 분석 결과 해석으로 틀린것은?
- bike.csv

```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_csv("bike.csv")
df.head(2)
```

```python
formula = "registered ~ season + holiday + season:holiday"
model = ols(formula, data = df).fit()
anova_lm(model)
```

```python
2.625279e-01
```

```python
2.785661e+04
```

```python
## 나는 틀림 C()안해줘서
```

정답

```python
formula = "registered ~ C(season) + C(holiday) + C(season):C(holiday)"
model = ols(formula, data = df).fit()
anova_lm(model)
```

### 2. 다음 중 종속변수를 registered, 독립변수를 season, weather로 지정했을 때 분석 결과 해석으로 틀린것은?
- bike.csv

```python
df = pd.read_csv("bike.csv")
df.head(2)
```

```python
df["weather"].unique()
```

```python
formula = "registered ~ C(season) + C(weather) + C(season):C(weather)"
model = ols(formula, data = df).fit()
anova_lm(model)
```

```python
anova_lm(model).round(5)
```

### 3. 다음 중 종속변수를 BMI, 독립변수를 연령대와 임신여부로 지정했을 때 교호작용 효과의 p-value는?
- diabetes.csv
- 나이가 70세 이상이거나 BMI가 0인 것은 제외하시오
- 임션여부는 Pregnancies가 0 초과인 경우 1, 나머지를 0으로 하시오
- 연령대는 10대, 20대 등 10살 단위로 구분하시오. ex) 10->1, 29->2

```python
df = pd.read_csv("diabetes.csv")
df.head(2)
```

```python
df2 = df.loc[(df["Age"] < 70) & (df["BMI"] > 0), ]
df2
```

```python
df2["imsin"] = (df["Pregnancies"] > 0) + 0
df2.head()
```

```python
df2["age_Dae"] = df["Age"] // 10
df2.head()
```

```python
formula = "BMI ~ C(age_Dae) + C(imsin) + C(age_Dae):C(imsin)"
model = ols(formula, data = df2).fit()
anova_lm(model).round(5)
```
