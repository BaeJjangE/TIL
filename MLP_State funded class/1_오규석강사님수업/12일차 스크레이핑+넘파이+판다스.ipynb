### meta 태그에서 인코딩 방식 추출하기

```python
# 정규 표현식 사용을 위해 re 모듈 import
import re

# 표준 에러를 지정하기 위해 sys 모듈을 import
import sys

# 웹 페이지 추츨을 위해 urlopen 모듈을 import
from urllib.request import urlopen

# 해당 홈페이지를 객체로 저장
f = urlopen("http://hanbit.co.kr/store/books/full_book_list.html")

# bytes 자료형을 변수에 저장
bytes_content = f.read()

# charset 은 HTML의 앞부분에 적혀 있는 경우가 많으므로 응답 본문의 앞부분 1024바이트를 ASCII문자로 디코딩
# ASCII 범위 이외의 문자는 U+FFFD로 변환되어 예외 발생 X
scanned_text = bytes_content[:1024].decode('ascii', errors='replace') # errors : 변환이 안 될 경우 응답을 지정

# 디코딩한 문자열에서 정규표현식으로 charset을 추출
# ["\']? => " or ' 를 포함하거나 포함하고 있지 않음
# () 소괄호는 묶어주는 역할
# ([\w-]+]) 안에서 -는 대시를 의미
# ([\w-]+]) => 알파벳 or 숫자 or _ or - 이 한번이상 반복됨
match = re.search(r'charset=["\']?([\w-]+])', scanned_text) # r이 문자열앞에 붙으면 해당 문자열이 구성된 그대로 문자열 반환 (순서변경x)
# r(row string), ex) print("aa\n") -> aa하고 띄어짐, print(r"aa\n") -> aa\n 으로 표시
# 'charset="utf-8'

# 인코딩 방식 중간확인
# print(match)

# 인코딩 방식을 변수에 저장
if match:
    encoding = match.group(1)
else:
    encoding = 'utf-8'

    
# 아래 print 구문이 에러로 출력
print('encoding :', encoding, file=sys.stderr) # file=sys.stderr는 print 함수의 형식

# 추출한 인코딩 방식으로 디코딩
text = bytes_content.decode(encoding)
print(text)
```

 - 'strict' = 에러가 발생하면 해당 에러를 보여줌
 - 'replace' = 디코딩한 방법의 범위 이외의 문자는 U+FFFD로 변환, 에러는 발생했으나 예외 처리됨
