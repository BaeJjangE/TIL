```python
DFList = []
for idx in range(Tournament.shape[0]):
    URL = requests.get(Tournament['telemetry_path'][idx])
    TextData = URL.text
    TextData_Json = json.loads(TextData)
    DFList.append(pd.DataFrame(TextData_Json))
    #DFList['map'] = Tournament['map_name'][idx]
```

```python
len(DFList)
```
